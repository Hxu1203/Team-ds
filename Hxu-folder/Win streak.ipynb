```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Import the dataset
fifa_df = pd.read_csv('./sport/fifa_world_cup.csv')
pd.options.display.max_columns = None
```

```python
#panda way to explore the correlation
correlation_series = fifa_df.corr()['home_team_score'].sort_values()
max_corr_value = correlation_series.iloc[-2]
max_corr_column = correlation_series.index[-2]
print("Most Positively Correlated Column:", max_corr_column)
print("Maximum Correlation Value:", max_corr_value)
```

# Winning Streak

```python
sorted_df= fifa_df.sort_values(by=(['home_team','date']), ascending=True)##create a dataframe that's sorted in order
sorted_df.loc[sorted_df["home_team_result"] == "Draw", "home_team_result"] ="Lose"##All the Draw consider as lose
sorted_df
```

# Filter out to only have World Cup data

```python
sorted_df = sorted_df.loc[(sorted_df['tournament'] == ('FIFA World Cup')),['home_team','tournament','home_team_result']]
sorted_df['tournament'].unique()
```

# Count the streaks

```python

sorted_df['start_of_streak'] = sorted_df.home_team_result.ne(sorted_df['home_team_result'].shift())##Identify the start of each streak
sorted_df['diff_team'] = sorted_df.home_team.ne(sorted_df['home_team'].shift())

sorted_df['start_of_streak'] = np.where((sorted_df['diff_team'] == True),True,sorted_df['start_of_streak'])## If it goes to different team, reset the streak
sorted_df['streak_id'] = sorted_df['start_of_streak'].cumsum()##
sorted_df['streak_counter'] = sorted_df.groupby('streak_id').cumcount()+1
sorted_df
sorted_df.head(30)
```
